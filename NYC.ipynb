التعامل مع YEAR BUILT
القيمة الشاذة هنا غالباً ما تكون رقم "0".

المشكلة: رقم 0 يعني أن سنة البناء غير معروفة.

الحل:

قم باستبدال الـ 0 بقيمة NaN ثم قرر: إما حذف هذه الصفوف، أو استبدالها بـ Median (الوسيط) لسنوات البناء في نفس المنطقة (BOROUGH).

أي سنة بناء قبل عام 1800 أو بعد 2026 تعتبر منطقياً خطأ بيانات ويجب استبعادها.

```python
print(df[df['YEAR BUILT']==0].count())
print ( df.shape[0])
```

```python
# 1671/32305 ~= 5%   its little data but we will not delete it but will prefer to use the median for each rigion to predit the year of built because this rows may cotains important informations
# استبدال الـ 0 بقيم فارغة أولاً

df['YEAR BUILT'] = df['YEAR BUILT'].replace(0, np.nan)

# ملء القيم الفارغة بوسيط السنة لكل منطقة على حدة
df['YEAR BUILT'] = df.groupby('BOROUGH')['YEAR BUILT'].transform(lambda x: x.fillna(x.median()))
```

```python
import plotly.express as px

# 1. أولاً: تحويل أرقام المناطق إلى أسمائها الحقيقية لتكون الخريطة واضحة
borough_map = {1: 'Manhattan', 2: 'Bronx', 3: 'Brooklyn', 4: 'Queens', 5: 'Staten Island'}
df['BOROUGH_NAME'] = df['BOROUGH'].map(borough_map)

# 2. تنظيف السعر من القيم الصفرية أو الصغيرة جداً (التي قد تفسد الرسم)
# سنركز على المبيعات الحقيقية (أكبر من 10,000 دولار مثلاً)
df_filtered = df[df['SALE PRICE'] > 10000]

# 3. رسم الـ Boxplot
fig = px.box(df_filtered, 
             x='BOROUGH_NAME', 
             y='SALE PRICE',
             color='BOROUGH_NAME', # تلوين كل منطقة بلون مختلف
             title='مقارنة أسعار العقارات بين مناطق نيويورك (Box Plot)',
             labels={'BOROUGH_NAME': 'المنطقة', 'SALE PRICE': 'سعر البيع'},
             points="outliers", # إظهار النقاط الشاذة فقط خارج الصندوق
             notched=True) # إضافة "فجوة" عند الوسيط لتمييز الفروق الإحصائية

# 4. تحسين شكل المحور الرأسي (لأن الأسعار قد تكون ضخمة جداً)
fig.update_layout(yaxis_type="log") # استخدام مقياس لوغاريتمي لرؤية أوضح إذا كانت الفروق ضخمة
fig.show()
```

## Time-Series Analysis

```python
import plotly.express as px

# 1. تجميع البيانات حسب السنة والشهر لحساب متوسط السعر
# سنستخدم الوسيط (Median) لأنه أكثر دقة في حالة وجود قيم شاذة
monthly_trend = df.groupby(['sale_year', 'sale_month'])['SALE PRICE'].median().reset_index()

# 2. دمج السنة والشهر في عمود واحد لتمثيل المحور الزمني بشكل صحيح
monthly_trend['period'] = monthly_trend['sale_year'].astype(str) + '-' + monthly_trend['sale_month'].astype(str).str.zfill(2)
monthly_trend = monthly_trend.sort_values('period')

# 3. رسم الخط الزمني
fig = px.line(monthly_trend, 
              x='period', 
              y='SALE PRICE',
              title='اتجاه السوق: تغير وسيط أسعار البيع شهرياً',
              labels={'period': 'الفترة (سنة-شهر)', 'SALE PRICE': 'وسيط سعر البيع ($)'},
              markers=True) # إضافة نقاط عند كل شهر لسهولة القراءة

# 4. تحسين مظهر الرسم
fig.update_layout(xaxis_tickangle=-45, template="plotly_white")
fig.show()
```

```python
import plotly.express as px

# 1. حساب عدد المبيعات لكل شهر
# نستخدم .size() لعد الصفوف (العمليات) في كل شهر
monthly_sales = df.groupby('sale_month').size().reset_index(name='sales_count')

# 2. تحويل أرقام الشهور إلى أسماء لسهولة القراءة
month_names = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 
               7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
monthly_sales['month_name'] = monthly_sales['sale_month'].map(month_names)

# 3. رسم مخطط الأعمدة
fig = px.bar(monthly_sales, 
             x='month_name', 
             y='sales_count',
             title='موسمية المبيعات: عدد الصفقات حسب شهر السنة',
             labels={'month_name': 'الشهر', 'sales_count': 'عدد العمليات المباعة'},
             color='sales_count', # تلوين الأعمدة حسب الكثافة
             color_continuous_scale='Viridis')

# 4. تحسين المظهر
fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':list(month_names.values())},
                  template="plotly_white")
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
import plotly.express as px

# 1. تصفية البيانات (استبعاد المباني التي سنة بنائها 0 أو غير منطقية)
df_age = df[(df['YEAR BUILT'] > 1850) & (df['SALE PRICE'] > 10000)].copy()

# 2. إنشاء عمود "العقد" (Decade)
df_age['Decade'] = (df_age['YEAR BUILT'] // 10 * 10).astype(int).astype(str) + 's'

# 3. حساب وسيط السعر لكل عقد (اختياري للتأكد من الأرقام)
decade_summary = df_age.groupby('Decade')['SALE PRICE'].median().sort_index()

# 4. رسم Boxplot للمقارنة
fig = px.box(df_age, 
             x='Decade', 
             y='SALE PRICE',
             color='Decade',
             title='أثر عمر المبنى: مقارنة أسعار العقارات حسب عقد البناء',
             labels={'Decade': 'عقد البناء', 'SALE PRICE': 'سعر البيع ($)'},
             category_orders={"Decade": sorted(df_age['Decade'].unique())}) # ترتيب العقود زمنياً

# 5. استخدام المقياس اللوغاريتمي لرؤية أوضح
fig.update_layout(yaxis_type="log", template="plotly_white", showlegend=False)
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
